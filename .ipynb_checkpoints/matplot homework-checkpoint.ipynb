Three observations: <br>

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load 
mouse_drug_data_to_load = "../Class Downloads/mouse_drug_data.csv"
clinical_trial_data_to_load = "../Class Downloads/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df=pd.read_csv(mouse_drug_data_to_load)
clinical_df=pd.read_csv(clinical_trial_data_to_load)
```

```python
# Combine the data into a single dataset
# Display the data table for preview
df=pd.merge(mouse_df, clinical_df, on="Mouse ID")
df.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint and convert to dataframe
mean_tumor_volume_df=pd.DataFrame(df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean().reset_index())
mean_tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
sem_tumor_volume_df=pd.DataFrame(df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem().reset_index())
sem_tumor_volume_df.head()
```

```python
#Reformat DataFrames
mean_tumor_volume_df=mean_tumor_volume_df.pivot_table(index='Timepoint', columns='Drug',values='Tumor Volume (mm3)')

sem_tumor_volume_df=sem_tumor_volume_df.pivot_table(index='Timepoint', columns='Drug',values='Tumor Volume (mm3)')

```

```python
mean_tumor_volume_df
```

```python
sem_tumor_volume_df.head()
```

```python
#Creating a scatter plot that shows how the tumor volume 
#changes over time for each treatment.
```

```python
fig, ax = plt.subplots()
for drug in mean_tumor_volume_df.columns:
    mean_tumor_volume_df.plot(kind='line', x=np.arange(0,50,5), y=drug, ax=ax, yerr=sem_tumor_volume_df[drug] ,figsize=(12,6))

plt.title("Tumor Response to Treatment")
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(True, axis='y')
plt.legend(loc='best')
plt.xlim(0,48)
plt.show()
```

```python
#Survival Rates
#Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
#Convert to DataFrame
```

```python
survival_df=pd.DataFrame(df.groupby(['Drug', 'Timepoint'])['Mouse ID'].count())
```

```python
survival_df=survival_df.pivot_table(index="Timepoint", columns="Drug", values="Mouse ID")
survival_df
```

```python
drug_names=survival_df.columns.tolist()
drug_names

survival_df.index
```

```python
survival_pct_df=pd.DataFrame(columns=drug_names, index= survival_df.index)
```

```python
for drug in drug_names:
    for i in np.arange(0,50,5):
        survival_pct_df.loc[i][drug]=survival_df.loc[i][drug]/survival_df.loc[0][drug]*100
        
```

```python
survival_pct_df
```

```python
#Plot
#Creating a scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)
```

```python
ax=plt.gca()
for drug in mean_tumor_volume_df.columns:
    survival_pct_df.plot(kind='line', x=np.arange(0,50,5), y=drug, ax=ax, figsize=(12,6))

plt.title("Survival During Treatment")
plt.xlabel('Time (Days)')
plt.ylabel('Survial Rate (%)')
plt.grid(True)
#plt.figure(figsize=(20,6), dpi=100)
plt.legend(loc='best')
plt.xlim(0,48)
plt.show()
```

```python
#Summary 
# Calculate the total % tumor volume change for each drug across the full 45 days.
# Display the data to confirm
```

```python
tumor_change_list=[]
for drug in mean_tumor_volume_df.columns:
    change=(mean_tumor_volume_df[drug][45]-mean_tumor_volume_df[drug][0])/mean_tumor_volume_df[drug][0]*100
    tumor_change_list.append([drug, change])


tumor_change=tuple(tumor_change_list)

tumor_change
```

```python
#Function to return average tumor volume change accross drugs

def getKey(item):
    return item[1]
```

```python
tumor_change=sorted(tumor_change, key=getKey)
```

```python
xlabels, yvalue = [*zip(*tumor_change)]
```

```python
x_axis=np.arange(len(tumor_change))
```

```python
#function to color bars based on tumor volumen change
def bar_color(x):
    if x >0:
        return 'r'
    else: 
        return 'g'

colors=[]
for x in yvalue:
    colors.append(bar_color(x))

colors
```

```python
#function to put values in bar chart

def autolabel(rects):
    for rect in rects:
        height=rect.get_height()
        ax.text(rect.get_x() + + rect.get_width()/2., 1.05*height, '%d' % int(height), ha='center', va='bottom')
        
```

```python
fig, ax = plt.subplots()

bar_plot=ax.bar(x_axis, yvalue, align="center", color=colors, alpha=0.8, width=1.0, edgecolor='black', linewidth=0.2)

tick_locations= [value for value in x_axis]
plt.xticks(tick_locations, xlabels, rotation=65)

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.grid(True, color='grey', linestyle='-', linewidth=0.25, alpha=0.75)
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(min(yvalue)-10, max(yvalue)+10)
plt.figure(figsize=(12,6))
autolabel(bar_plot)

plt.show()
```
